## 2021

```python
q1_path = '/content/drive/MyDrive/CHINA_HATE/2022/2022_Q1.csv'
q1 = pd.read_csv('/content/drive/MyDrive/CHINA_HATE/2022/2022_Q1.csv')
```

```python
q2_path = '/content/drive/MyDrive/CHINA_HATE/2022/2022_Q2.csv'
q2 = pd.read_csv('/content/drive/MyDrive/CHINA_HATE/2022/2022_Q2.csv', lineterminator='\n')
```

```python
q3_path = '/content/drive/MyDrive/CHINA_HATE/2022/2022_Q3.csv'
q3 = pd.read_csv('/content/drive/MyDrive/CHINA_HATE/2022/2022_Q3.csv')
```

```python
q4_path = '/content/drive/MyDrive/CHINA_HATE/2022/2022_Q4.csv'
q4 = pd.read_csv('/content/drive/MyDrive/CHINA_HATE/2022/2022_Q4.csv', lineterminator='\n')
```

```python
df_2022 = pd.concat([q1, q2, q3, q4], axis=0)
```

```python
df_na = df_2022[df_2022.isna().any(axis=1)]
```

```python
df_2022 = df_2022.fillna(0)
```

```python
df_2022[['CARDIFFNLP_ROBERTA_PREDS', 'hateXplain_PREDS', 'COVID_HATE_BERT_PREDS']] = df_2022[['CARDIFFNLP_ROBERTA_PREDS', 'hateXplain_PREDS', 'COVID_HATE_BERT_PREDS']].astype(int)
```

```python
df_2022
```

```python
Counter(df_2022['COVID_HATE_BERT_PREDS'])
```

```python
df_2022['COVID_HATE_BERT_PREDS']
```

```python
len_df = len(df_2022)
COVID_HATE_BERT_PREDS_bin = []
hateXplain_PREDS_bin = []

covid_hate_preds = list(df_2022['COVID_HATE_BERT_PREDS'])
hatexplain_preds = list(df_2022['hateXplain_PREDS'])

for i in range(len_df):
  if covid_hate_preds[i] == 2:
    COVID_HATE_BERT_PREDS_bin.append(1)
  else:
    COVID_HATE_BERT_PREDS_bin.append(0)
  
  if hatexplain_preds[i] == 0:
    hateXplain_PREDS_bin.append(1)
  else:
    hateXplain_PREDS_bin.append(0)

df_2022['COVID_HATE_BERT_PREDS_bin'] = COVID_HATE_BERT_PREDS_bin
df_2022['hateXplain_PREDS_bin'] = hateXplain_PREDS_bin
```

```python
voting_2022 = []

cardiff_preds = list(df_2022['CARDIFFNLP_ROBERTA_PREDS'])

for i in range(len_df):

  votes = []
  votes.append(COVID_HATE_BERT_PREDS_bin[i])
  votes.append(cardiff_preds[i])
  votes.append(hateXplain_PREDS_bin[i])

  voting_2022.append(max(set(votes), key = votes.count))

df_2022['voting_results'] = voting_2022
Counter(voting_2022)
```

```python
df_2022_hate = df_2022.loc[df_2022['voting_results']==1]
df_2022_hate
```

```python
samples_2022 = df_2022_hate.sample(n=800)
samples_2022
```

```python
samples_2022.to_csv('samples_2022.csv', index=False)
# np.savetxt(r'samples_2020.txt', samples_2020.tweet, fmt='%d')
```
