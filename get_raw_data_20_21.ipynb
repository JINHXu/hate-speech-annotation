## 2020

```python
df_2020 = pd.read_csv(path_2020)
```

```python
len_df = len(df_2020)
COVID_HATE_BERT_PREDS_bin = []
hateXplain_PREDS_bin = []

for i in range(len_df):
  if df_2020['COVID_HATE_BERT_PREDS'][i] == 2:
    COVID_HATE_BERT_PREDS_bin.append(1)
  else:
    COVID_HATE_BERT_PREDS_bin.append(0)
  
  if df_2020['hateXplain_PREDS'][i] == 0:
    hateXplain_PREDS_bin.append(1)
  else:
    hateXplain_PREDS_bin.append(0)

df_2020['COVID_HATE_BERT_PREDS_bin'] = COVID_HATE_BERT_PREDS_bin
df_2020['hateXplain_PREDS_bin'] = hateXplain_PREDS_bin
```

```python
voting_2020 = []

for i in range(len_df):

  votes = []
  votes.append(df_2020['COVID_HATE_BERT_PREDS_bin'][i])
  votes.append(df_2020['CARDIFFNLP_ROBERTA_PREDS'][i])
  votes.append(df_2020['hateXplain_PREDS_bin'][i])

  voting_2020.append(max(set(votes), key = votes.count))

df_2020['voting_results'] = voting_2020
Counter(voting_2020)
```

```python
df_2020_hate = df_2020.loc[df_2020['voting_results']==1]
df_2020_hate
```

```python
# from random import sample
  
# tweets_2020 = list(df_2020_hate['tweet'])
# samples = sample(tweets_2020, 800)
```

```python
samples_2020 = df_2020_hate.sample(n=800)
samples_2020
```

```python
samples_2020.to_csv('samples_2020.csv', index=False)
# np.savetxt(r'samples_2020.txt', samples_2020.tweet, fmt='%d')
```

```python
len_df = len(df_2021)
COVID_HATE_BERT_PREDS_bin = []
hateXplain_PREDS_bin = []

for i in range(len_df):
  if df_2021['COVID_HATE_BERT_PREDS'][i] == 2:
    COVID_HATE_BERT_PREDS_bin.append(1)
  else:
    COVID_HATE_BERT_PREDS_bin.append(0)
  
  if df_2021['hateXplain_PREDS'][i] == 0:
    hateXplain_PREDS_bin.append(1)
  else:
    hateXplain_PREDS_bin.append(0)

df_2021['COVID_HATE_BERT_PREDS_bin'] = COVID_HATE_BERT_PREDS_bin
df_2021['hateXplain_PREDS_bin'] = hateXplain_PREDS_bin
```
